```python
# Import statements
import os
import numpy as np

import xml.etree.ElementTree as ET
from PIL import Image, ImageDraw
```

```python
images = [(os.path.join(data_path, f), f) for f in os.listdir(data_path) if (f.endswith('.jpg'))]
xml = [(os.path.join(data_path, f), f) for f in os.listdir(data_path) if (f.endswith('.xml'))]

images.sort()
xml.sort()

assert(len(images) == len(xml))

joined_list = []
for i in range(len(images)):
  joined_list.append((images[i], xml[i]))

print(joined_list)
```

```python
# create a list of all the unique module names to assign them a unique colour

all_modules = []

for path, _ in xml:
  tree = ET.parse(path)
  root = tree.getroot()

  for module in root.iter('name'):
    if module.text not in all_modules:
      all_modules.append(module.text)

print(all_modules)
print(len(all_modules))
```

```python
# perform the unique colour assignment

colours = [
        "#FF0000", "#FF7F50", "#B0171F", "#872657", "#FF00FF",
        "#FFFF00", "#FF8000", "#FF9912", "#DAA569", "#FF6100",
        "#0000FF", "#3D59AB", "#03A89E", "#33A1C9", "#00C78C",
        "#00FF00", "#385E0F", "#00C957", "#6B8E23", "#2E8B57",
        "#A020F0", "#8A2BE2", "#A066D3", "#DA70D6", "#DDA0DD",
        '#E6194B', '#3CB44B', '#FFE119', '#4363D8', '#F58231',
        '#911EB4', '#46F0F0', '#F032E6', '#BCF60C', '#FABEBE',
        '#008080', '#E6BEFF', '#9A6324', '#FFFAC8', '#800000',
        '#AAFFC3', '#808000', '#FFD8B1', '#000075', '#808080']

colour_assignment = {}
for idx, m in enumerate(all_modules):
  colour_assignment[m] = colours[idx]

print(colour_assignment)
```

```python
upload_path = "/content/drive/MyDrive/APS360_Images/BoundingBoxes/"

for image, xml in joined_list:
  image_path, image_name = image
  xml_path, xml_name = xml
  
  tree = ET.parse(xml_path)
  root = tree.getroot()
  
  annotations = []

  for obj in root.findall('object'):
    module = obj.find('name').text
    bndbox = obj.find('bndbox')
    xmin = int(bndbox.find('xmin').text)
    ymin = int(bndbox.find('ymin').text)
    xmax = int(bndbox.find('xmax').text)
    ymax = int(bndbox.find('ymax').text)
    annotations.append((module, [xmin, ymin, xmax, ymax]))

  image_drawn = Image.open(image_path).copy()

  img_bbox = ImageDraw.Draw(image_drawn)

  for module, bbox in annotations:
    img_bbox.rectangle(bbox, fill=colour_assignment[module], outline=colour_assignment[module]) 
    
  image_drawn.save(upload_path + image_name)
```
